# Read in Frequencies from File

```python
import pandas as pd
df = pd.read_csv("background_freqs.csv")
df 
```

# Generate Random Sequences using those frequencies

```python
n = 517**2 #(89+94)**2 # total number of random sequences to generate


# pull all of the frequencies from the file into a list to iterate through
cols = list(df.columns)
probs = []
for c in cols:
    probs.append(list(df[c]))

# iterate through probabilities at each base and generate random base for each position based on its probability
# then combine the list of bases into a single seqeunce
# repeat this process n times to generate n sequences
import numpy as np
random_seq_ls = [] # holds all of the random sequences
for i in range(0,n):
    random_seq = [] # holds a single random sequence 
    for p in probs:
        random_seq.append(np.random.choice(['A', 'C', 'G', 'U'], p=p))

    random_seq_ls.append(''.join(random_seq))

```

```python
rand_df = pd.DataFrame(random_seq_ls)
```

```python
rand_df.to_csv("random_sequences_"+str(n)+".csv")
```
